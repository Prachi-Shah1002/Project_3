```python
#Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from pymongo import MongoClient
```

```python
#CSV File to Load

NYC_data = Path("Data/NYC_Accidents_2020.csv")

#Read the CSV File
NYC = pd.read_csv(NYC_data,delimiter = ",")

#Display the data
NYC.head()
```

```python
#Checking for missing Data
NYC.count()
```

```python
# Clean the Data by Making a New Dataframe 

NYC_df = NYC.loc[:,["CRASH DATE","CRASH TIME","BOROUGH","ZIP CODE","LATITUDE","LONGITUDE","COLLISION_ID","VEHICLE TYPE CODE 1"]]
NYC_df.head()
```

```python
#Checking for missing Data
NYC_df.count()
```

```python
# Delete the Null values from row

NYC_df.dropna(subset=['BOROUGH','ZIP CODE','LATITUDE','LONGITUDE','VEHICLE TYPE CODE 1'], inplace=True)
```

```python
NYC_df.count()
```

```python
NYC_df.head()
```

```python
# Rename the Columns Name

NYC_data_df = NYC_df.rename(columns={"CRASH DATE":"Crash_Date","CRASH TIME":"Crash_Time","BOROUGH":"City_Name",
                                    "ZIP CODE":"Zip_Code","LATITUDE":"Latitude","LONGITUDE":"Longitude",
                                    "COLLISION_ID":"Collision_id","VEHICLE TYPE CODE 1":"Vechicle_Type"})

NYC_data_df.head()
```

```python
# Export the NYC_data_df as CSV file.

NYC_data_df.to_csv("Data/NYC_Data.csv", index = False)
```
